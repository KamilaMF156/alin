# Transformasi Linear

## Definisi

Transformasi linier adalah fungsi dari satu ruang vektor ke ruang vektor lainnya yang mengikuti struktur dasar (linier). Transformasi linier juga dikenal sebagai operator linier atau pemetaan.

Pemetaan $T$ adalah Transformasi linier, jika:

1. $T(\mathbf{u}+\mathbf{v}) = T(\mathbf{u})+T(\mathbf{v})$
> Pembuktian:

Diberikan fungsi $T : R^2 → R^2$ yang dalam hal ini T(x,y) = (2x, y), akan
ditunjukkan bahwa T adalah transformasi linier. Misalkan u dan v adalah dua buah vektor di $R^2$, $u = (u_1,u_2)$ dan $v=(v_1,v_2)$. Maka,

$\begin{align}
T(\mathbf{u}+\mathbf{v}) &= T(u_1+v_1,u_2+v_2) \\ &= (2(u_1 + v_1), u_2 + v_2) \\ &= (2u_1 + 2v_1, u_2 + v_2) \\ &= \left[\begin{array}{c} 2u_1+2v_1 \\ u_2+v_2 \end{array} \right] \\ &= \left[\begin{array}{c} 2u_1 \\ u_2 \end{array} \right] + \left[\begin{array}{c} 2v_1 \\ v_2 \end{array} \right] \\ &= T(\mathbf{u}) + T(\mathbf{v})
\end{align}$

<br><br>

2. $\displaystyle T(\alpha \mathbf{v}) = \alpha T(\mathbf{v})$
> Pembuktian:

Diberikan fungsi $T : R^2 → R^2$ yang dalam hal ini T(x,y) = (2x, y), akan
ditunjukkan bahwa T adalah transformasi linier. Misalkan u dan v adalah dua buah vektor di $R^2$, $u = (u_1,u_2)$ dan $v=(v_1,v_2)$. Maka,

$\begin{align}
T(\alpha \mathbf{v}) &= T(\alpha v_1,\alpha v_2) \\ &= (2\alpha v_1,\alpha v_2) \\ &= \left[\begin{array}{c} 2 \alpha v_1 \\ \alpha v_2 \end{array} \right] \\ &= \alpha \left[\begin{array}{c} 2v_1 \\ v_2 \end{array} \right] \\ &= \alpha T(\mathbf{v})
\end{align}$

## Contoh Soal

Misalkan $T$ adalah transformasi linier $T:R^2 → R^2 \text { dimana}$

$
T\left[\begin{array}{c} 2 \\ 0 \end{array} \right] = \left[\begin{array}{c} -2 \\ 0 \end{array} \right], T\left[\begin{array}{c} 2 \\ 2 \end{array} \right] = \left[\begin{array}{c} -2 \\ 2 \end{array} \right]
$

Carilah matriks $A$!

<br><br>

**Solusi**

> $
\left[\begin{array}{c} 1 \\ 0 \end{array} \right] = x\left[\begin{array}{c} 2 \\ 0 \end{array} \right] + y\left[\begin{array}{c} 2 \\ 2 \end{array} \right]
$

$0=2y \\ 1=2x+2y \\ 1=2x+0 \\ 1=2x \\ \frac 12=x$

Hasil persamaan linier di atas akan dimasukkan ke

$T
\left[\begin{array}{c} 1 \\ 0 \end{array} \right] = x\left[\begin{array}{c} -2 \\ 0 \end{array} \right] + y\left[\begin{array}{c} -2 \\ 2 \end{array} \right]
$

Maka akan didapatkan kolom pertama matriks A,

$T
\left[\begin{array}{c} 1 \\ 0 \end{array} \right] = \frac{1}{2} \left[\begin{array}{c} -2 \\ 0 \end{array} \right] + 0\left[\begin{array}{c} -2 \\ 2 \end{array} \right] = \left[\begin{array}{c} -1 \\ 0 \end{array} \right]
$

<br>
<br>

> $
\left[\begin{array}{c} 0 \\ 1 \end{array} \right] = x\left[\begin{array}{c} 2 \\ 0 \end{array} \right] + y\left[\begin{array}{c} 2 \\ 2 \end{array} \right]
$

$1=2y \\ \frac{1}{2} = y \\  0=2x+2y \\ 0=2x+1 \\ -1=2x \\ -\frac{1}{2}=x$

Hasil persamaan linier di atas akan dimasukkan ke

$T
\left[\begin{array}{c} 0 \\ 1 \end{array} \right] = x\left[\begin{array}{c} -2 \\ 0 \end{array} \right] + y\left[\begin{array}{c} -2 \\ 2 \end{array} \right]
$

Sehingga akan didapatkan kolom kedua matriks A,

$T
\left[\begin{array}{c} 1 \\ 0 \end{array} \right] = (-\frac{1}{2}) \left[\begin{array}{c} -2 \\ 0 \end{array} \right] + \frac{1}{2} \left[\begin{array}{c} -2 \\ 2 \end{array} \right] = \left[\begin{array}{c} 1 \\ 0 \end{array} \right] + \left[\begin{array}{c} -1 \\ 1 \end{array} \right] = \left[\begin{array}{c} 0 \\ 1 \end{array} \right]
$

<br>

Jadi, matriks A adalah $
\begin{bmatrix} -1 & 0 \\ 0 & 1 \end{bmatrix}
$

## Refleksi pada Transformasi

Refleksi atau pencerminan adalah transformasi yang memindahkan setiap titik pada suatu bidang dengan menggunakan sifat bayangan pada cermin. Refleksi merupana salah satu jenis transformasi yang mana benda yang direfleksikan biasanya berupa titik, kurva, bangun datar, atau bangun ruang dengan garis sebagai 'cermin'-nya.

Sifat-sifat:
- Objek yang direfleksikan tidak mengalami perubahan bentuk dan ukuran.
- Jarak objek dengan cermin(datar) sama dengan jarak bayangan dengan cermin.

### Rumus-rumus

**2 Dimensi**

1. Pencerminan terhadap titik pusat koordinat O(0,0)

$A(x,y) ⟶ A'(x',y')$

$(x,y) ⟶ (-x,-y)$

$\begin{pmatrix} x' \\ y' \end{pmatrix} = \begin{pmatrix} -1 & 0 \\ 0 & -1 \end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix}$

2. Pencerminan terhadap sumbu x

$A(x,y) ⟶ A'(x',y')$

$(x,y) ⟶ (x,-y)$

$\begin{pmatrix} x' \\ y' \end{pmatrix} = \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix}$

3. Pencerminan terhadap sumbu y

$A(x,y) ⟶ A'(x',y')$

$(x,y) ⟶ (-x,y)$

$\begin{pmatrix} x' \\ y' \end{pmatrix} = \begin{pmatrix} -1 & 0 \\ 0 & 1 \end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix}$

4. Pencerminan terhadap garis y=x

$A(x,y) ⟶ A'(x',y')$

$(x,y) ⟶ (y,x)$

$\begin{pmatrix} x' \\ y' \end{pmatrix} = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix}$

5. Pencerminan terhadap garis y=-x

$A(x,y) ⟶ A'(x',y')$

$(x,y) ⟶ (-y,-x)$

$\begin{pmatrix} x' \\ y' \end{pmatrix} = \begin{pmatrix} 0 & -1 \\ -1 & 0 \end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix}$

<br>

**3 Dimensi**
1. Pencerminan terhadap bidang xy

$A(x,y,z) ⟶ A'(x',y',z')$

$(x,y,z) ⟶ (x,y,-z)$

$\begin{pmatrix} x' \\ y' \\ z' \end{pmatrix} = \begin{pmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & -1 \end{pmatrix} \begin{pmatrix} x \\ y \\ z \end{pmatrix}$

2. Pencerminan terhadap bidang xz

$A(x,y,z) ⟶ A'(x',y',z')$

$(x,y,z) ⟶ (x,-y,z)$

$\begin{pmatrix} x' \\ y' \\ z' \end{pmatrix} = \begin{pmatrix} 1 & 0 & 0 \\ 0 & -1 & 0 \\ 0 & 0 & 1 \end{pmatrix} \begin{pmatrix} x \\ y \\ z \end{pmatrix}$

3. Pencerminan terhadap bidang yz

$A(x,y,z) ⟶ A'(x',y',z')$

$(x,y,z) ⟶ (-x,y,z)$

$\begin{pmatrix} x' \\ y' \\ z' \end{pmatrix} = \begin{pmatrix} -1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{pmatrix} \begin{pmatrix} x \\ y \\ z \end{pmatrix}$

### Implementasi numpy

```python
import numpy as np

# Implementasi 2 Dimensi
sumbuO = np.array([[-1,0],[0,-1]])
matriksA = np.array([[-1,0],[0,1]])
refleksi2D = np.dot(sumbuO, matriksA)
print("Matriks Awal :")
print(matriksA)
print("Hasil Refleksi :")
print(refleksi2D)

print()

# Implementasi 3 Dimensi
bidangYZ = np.array([[-1,0,0],[0,1,0],[0,0,1]])
matriksB = np.array([[5,0,1],[2,1,1],[3,1,0]])
refleksi3D = np.dot(bidangYZ, matriksB)
print("Matriks Awal :")
print(matriksB)
print("Hasil Refleksi :")
print(refleksi3D)
```
